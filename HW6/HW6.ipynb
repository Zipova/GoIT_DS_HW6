```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('data_2d.csv', header=None)
data2d = data.iloc[:, [1, 2]]
sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(data2d)
    sse[k] = kmeans.inertia_ 
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
kmeans2d = KMeans(n_clusters=2, n_init='auto', max_iter=1000).fit(data2d)
centroids = kmeans2d.cluster_centers_
plt.scatter(data2d.iloc[:, [0]], data2d.iloc[:, [1]], c=kmeans2d.labels_)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red')
plt.show()
```
